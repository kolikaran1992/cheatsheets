# Heaps

## Definition
- **Binary heap** = nearly complete binary tree (filled leftâ†’right).

## Types
- **Max heap:** parent â‰¥ children  
- **Min heap:** parent â‰¤ children  

## Uses
- Max heap â†’ heap sort  
- Min heap â†’ priority queues  

## Complexity
- Height = O(log n); insert/delete = O(log n)

## Array Representation (1-indexed)
- Left child: `2*i`  
- Right child: `2*i + 1`  
- Parent: `i//2`
- heap size = len(arr) - 1


### **A node can NEVER have a right child without a left child.**

**Proof (one line):**
If `2*i + 1 â‰¤ n` (right exists), then `2*i â‰¤ nâˆ’1 < n`, so `2*i â‰¤ n` (left exists).


### ðŸŒ¿ **Internal nodes and leaves**

* **Internal nodes:** indices `1 â€¦ âŒŠn/2âŒ‹`
* **Leaves:** indices `âŒŠn/2âŒ‹ + 1 â€¦ n`
  (because for `i > âŒŠn/2âŒ‹`, `2*i > n` â†’ no children)


* Binary heap â‰¡ no missing inner children; all missing nodes appear only at the **last level**, from **right to left**.
---


## Example (max heap)
Array: `[ -, 21, 15, 18, 12, 13, 10, 9 ]`

# Build Max Heap

## 1. Max-Heapify
- Maintains **max-heap property** (parent â‰¥ children).
- Assumes children of index `i` are already heaps.
- If `A[i]` < one of its children â†’ swap with larger child â†’ recurse.
- Time: **O(log n)**.

**Pseudocode**
```

max_heapify(A, i):
l = 2*i
r = 2*i + 1
largest = i
if l â‰¤ heap_size and A[l] > A[i]: largest = l
if r â‰¤ heap_size and A[r] > A[largest]: largest = r
if largest != i:
swap(A[i], A[largest])
max_heapify(A, largest)

```

---

## 2. Build-Max-Heap
- Calls max-heapify starting from the **last non-leaf** down to the root.
- Leaves (indices `n/2 + 1` to `n`) are already heaps.
- Time: **O(n)**.

why **O(n)**? Because, leaves are already heaps and we start from parents of all the leaves which means heapify takes constant time

**Pseudocode**
```

build_max_heap(A):
heap_size = len(A)
for i in range(floor(n/2), 0, -1):
max_heapify(A, i)

```

## 3. Key Ideas
- Bottom-up approach â†’ fewer heapify calls.
- Useful for **heap sort** and priority queue building.

```

````python
def max_heapify(arr, n, i):
    largest = i
    left = 2 * i
    right = 2 * i + 1

    if left <= n and arr[left] > arr[largest]:
        largest = left

    if right <= n and arr[right] > arr[largest]:
        largest = right

    print(arr, "root =", i, "| largest index =", largest)

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        max_heapify(arr, n, largest)


def build_max_heap(arr):
    n = len(arr) - 1
    for i in range(n // 2, 0, -1):
        max_heapify(arr, n, i)
````

# Remove and Insert
## ðŸ”º INSERT (push)

1. **Place new value at the next open leaf**
   (heaps fill **left â†’ right** on the last level).
2. **Bubble Up (Swap Up)**
   While `value > parent`:
   â†’ **swap with parent**.

**Goal:** restore *parent â‰¥ children* property.

---

## ðŸ”» REMOVE (pop)

(Removing from a **max heap**, so we remove the root.)

1. **Swap root with last leaf.**
2. **Remove the last leaf** (this is the max value).
3. **Bubble Down (Heapify Down)**
   At each step:

   * Compare parent with *its two children*.
   * **Pick the larger child**.
   * If `child > parent`: swap.

**Goal:** restore *parent â‰¥ children* property by pushing the small value down.

---

# â­ One-line summaries

* **Insert:** place at bottom â†’ swap up until parent is bigger.
* **Remove:** swap with last â†’ remove â†’ swap down with larger child until heap is valid.

---

````python
import math


class MaxHeap:
    def __init__(self, arr=[]):
        self.heap = [None] + arr
        self._build()

    @property
    def size(self):
        return len(self.heap) - 1

    def _heapify(self, i) -> None:
        # children nodes
        left = 2 * i
        right = 2 * i + 1

        # initialise largest as root
        largest = i

        # see if left child of root exists and is greater than root
        if left <= self.size and self.heap[left] > self.heap[largest]:
            largest = left

        # see if right child of root exists and is greater than root
        if right <= self.size and self.heap[right] > self.heap[largest]:
            largest = right

        # swap and continue heapifying if root is not largest
        if largest != i:
            self.heap[i], self.heap[largest] = self.heap[largest], self.heap[i]
            self._heapify(largest)

    def _build(self) -> None:
        for i in range(self.size // 2, 0, -1):
            self._heapify(i)

    def __repr__(self):
        if self.size == 0:
            return "<empty heap>"

        import math

        total_height = int(math.log2(self.size)) + 1
        out = []

        for level in range(total_height):
            start = 2**level
            end = min(2 ** (level + 1), self.size + 1)

            nodes = [str(self.heap[i]) for i in range(start, end)]

            # center spacing per tree level
            gap = " " * (2 ** (total_height - level - 1))

            out.append(gap + gap.join(nodes))

        return "\n".join(out)

    def insert(self, key) -> None:
        self.heap.append(key)
        index = self.size
        parent = index // 2

        # bubble up
        while index > 1:
            parent = index // 2
            if self.heap[parent] < self.heap[index]:
                self.heap[parent], self.heap[index] = (
                    self.heap[index],
                    self.heap[parent],
                )
                index = parent
            else:
                break

    def pop_max(self):
        if self.size == 0:
            raise IndexError("pop from empty heap")

        max_value = self.heap[1]
        self.heap[1] = self.heap[self.size]
        self.heap.pop()
        # heapify down from the root
        self._heapify(1)
        return max_value
````

# ðŸŒ³ **Nearly Complete Binary Tree Height Cheatsheet**

## **1. Height from the TOP (root-based height)**

* Level numbering: root = level 0
* Height = maximum number of edges from root to any leaf

### **Formula**

$$
H_{\text{top}} = \lfloor \log_2(n) \rfloor
$$

### **Nodes at level ( $\ell$ )**

$$
\text{nodes at level } \ell = 2^\ell
$$

---

## **2. Height from the BOTTOM (subtree height)**

Used for heapify-down cost analysis.

* Leaves = height 0
* Their parents = height 1
* Root = height ($ H_{\text{top}} $)

### **Formula**

$$
H_{\text{bottom}} = H_{\text{top}} = \lfloor \log_2(n) \rfloor
$$

### **Nodes with subtree height ( h )**

$$
\text{nodes with height } h \approx \frac{n}{2^{h+1}}
$$

---

# ðŸ“ **Quick Mapping Between the Two**

| Top height (distance from root) | Bottom height (distance from leaf) | Approx # nodes |
| ------------------------------- | ---------------------------------- | -------------- |
| 0                               | H                                  | 1              |
| 1                               | Hâˆ’1                                | 2              |
| 2                               | Hâˆ’2                                | 4              |
| â€¦                               | â€¦                                  | â€¦              |
| H                               | 0                                  | n/2            |

---

# â­ **Most Important Identity for HEAP Analysis**

$$
\sum_{h=0}^{H} h \cdot \frac{n}{2^{h+1}} = O(n)
$$

Thatâ€™s why **BUILD-HEAP is O(n)**.

---

````python
class FixedSizeMinHeap:
    def __init__(self, max_size: int = 3):
        self.heap = [None]
        self.max_size = max_size

    @property
    def size(self):
        return len(self.heap) - 1

    def peek(self) -> int:
        if self.size == 0:
            raise IndexError("peek from empty heap")
        return self.heap[1]

    def _base_insert(self, val: int) -> None:
        self.heap.append(val)
        index = self.size
        parent = index // 2

        # bubble up
        while index > 1:
            parent = index // 2
            if self.heap[parent] > self.heap[index]:
                self.heap[parent], self.heap[index] = (
                    self.heap[index],
                    self.heap[parent],
                )
                index = parent
            else:
                break

    def _base_pop(self) -> int:
        if self.size == 0:
            raise IndexError("pop from empty heap")

        min_value = self.heap[1]
        self.heap[1] = self.heap[self.size]
        self.heap.pop()
        # heapify down from the root
        self._heapify_down(1)
        return min_value

    def _heapify_down(self, i) -> None:
        # children nodes
        left = 2 * i
        right = 2 * i + 1

        # initialise smallest as root
        smallest = i

        # see if left child of root exists and is smaller than root
        if left <= self.size and self.heap[left] < self.heap[smallest]:
            smallest = left

        # see if right child of root exists and is smaller than root
        if right <= self.size and self.heap[right] < self.heap[smallest]:
            smallest = right

        # swap and continue heapifying if root is not smallest
        if smallest != i:
            self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
            self._heapify_down(smallest)

    def insert(self, val: int) -> None:
        if self.size < self.max_size:
            self._base_insert(val)
        else:
            if val >= self.heap[1]:
                self._base_pop()
                self._base_insert(val)

    def __repr__(self):
        if self.size == 0:
            return "<empty heap>"

        import math

        total_height = int(math.log2(self.size)) + 1
        out = []

        for level in range(total_height):
            start = 2**level
            end = min(2 ** (level + 1), self.size + 1)

            nodes = [str(self.heap[i]) for i in range(start, end)]

            # center spacing per tree level
            gap = " " * (2 ** (total_height - level - 1))

            out.append(gap + gap.join(nodes))

        return "\n".join(out)
````
